### Generate Dataset

```python
%matplotlib inline
from IPython import display
from matplotlib import pyplot as plt
from mxnet import autograd, nd
import random
```

$y = Xw+b+e$ with $e$~$N(0,0.01)$

```python
num_inputs = 2
num_examples = 1000
true_w = [2,-3.4]
true_b = 4.2
features = nd.random.normal(scale = 1, shape = (num_examples, num_inputs))
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b
labels += nd.random.normal(scale = 0.01, shape = labels.shape)
```

### Read data

```python
# return batch size data from inputs
def data_iter(batch_size, features,labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0,num_examples,batch_size):
        j = nd.array(indices[i:min(i + batch_size,num_examples)])
        yield features.take(j), labels.take(j)
```

```python
batch_size = 10
for X,y in data_iter(batch_size,features,labels):
    print(X,y)
    break
```

### Randomize the initial parameter

```python
w = nd.random.normal(scale = 0.01, shape = (num_inputs,1))
b = nd.zeros(shape=(1,))
w.attach_grad()
b.attach_grad()
```

### Define Model

```python
def linreg(X,w,b):
    return nd.dot(X,w)+b
```

### Define the loss function

```python
def squared_loss(y_hat,y):
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

### Define the optimize function

```python
def sgd(params,lr,batch_size):
    for param in params:
        param[:] = param - lr*param.grad / batch_size
```

### Train your model

```python
lr = 0.03
num_epochs = 5
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    for X,y in data_iter(batch_size,features,labels):
        with autograd.record():
            l = loss(net(X,w,b),y)
        l.backward()
        sgd([w,b],lr,batch_size)
    train_l = loss(net(features,w,b),labels)
    print("epoch %d, loss %f" %(epoch + 1, train_l.mean().asnumpy()))
```

## Linear regression with `mxnet`

### Generate data

```python
from mxnet import autograd,nd

num_inputs = 2
num_examples = 1000
true_w = [2,-3.4]
true_b = 4.2
features = nd.random.normal(scale = 1, shape = (num_examples, num_inputs))
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b
labels += nd.random.normal(scale = 0.01, shape = labels.shape)
```

### Read data

```python
from mxnet.gluon import data as gdata

batch_size = 10
dataset = gdata.ArrayDataset(features,labels)
# get the batch size of data
data_iter = gdata.DataLoader(dataset,batch_size,shuffle = True)
```

```python
for X,y in data_iter:
    print(X,y)
    break
```

### Define the model
* fully-connected model (`Dense`)
* only one output

```python
from mxnet.gluon import nn

net = nn.Sequential()
net.add(nn.Dense(1))
```

### Initialize the paramter
* bias would be assigned as 0 automatically

```python
from mxnet import init

net.initialize(init.Normal(sigma = 0.01))
```

### Define the loss function
* we use $l_2$ norm here

```python
from mxnet.gluon import loss as gloss
loss = gloss.L2Loss()
```

### Define the optimize function
* use the instance of `Trainer`

```python
from mxnet import gluon
```

```python
trainer = gluon.Trainer(net.collect_params(),'sgd',{'learning_rate':0.03})
```

### Train the model

```python
net
```

```python
num_epochs = 5
for epoch in range(1,num_epochs+1):
    for X,y in data_iter:
        with autograd.record():
            l = loss(net(X),y)
        l.backward()
        trainer.step(batch_size) # point out the batch size
    l = loss(net(features),labels)
    print('epch %d, loss: %f' %(epoch,l.mean().asnumpy()))
        
```

## `softmax` regression


So, we have $y_1 + y_2 + y_3 = 1$ and $ 0<=y_1,y_2,y_3<=1$


### get the data

```python
%matplotlib inline
import d2lzh as d2l
from mxnet import autograd,nd
```

```python
batch_size = 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

### Initialize model parameters
* 28 * 28 = 784
* 10 labels

```python
num_inputs = 784
num_outputs = 10

W = nd.random.normal(scale = 0.01,shape=(num_inputs,num_outputs))
b = nd.zeros(num_outputs)
```

```python
W.attach_grad()
b.attach_grad()
```

### Realize `SoftMax`

```python
X = nd.array([[1,2,3],[4,5,6]])
X
```

```python
X.sum(axis = 0,keepdims = True),X.sum(axis = 1,keepdims = True)
```

```python
X.sum(axis = 0),X.sum(axis = 1)
```

```python
def softmax(X):
    x_exp = X.exp()
    partition = x_exp.sum(axis = 1,keepdims = True)
    return x_exp /partition
```

```python
X = nd.random.normal(shape = (20,5))
X_prob = softmax(X)
X_prob,X_prob.sum(axis = 1)
```

### Define Model

```python
def net(X):
    return softmax(nd.dot(X.reshape((-1,num_inputs)),W)+b)
```

### Cross entropy


### Define loss function
* `nd.pick()` function returns the element from first array by second array(as index)

```python
y_hat = nd.array([[0.1,0.3,0.6],[0.3,0.2,0.5]])
y = nd.array([0,2],dtype = 'int32')
nd.pick(y_hat,y)
```

```python
def cross_entropy(y_hat,y):
    return -nd.pick(y_hat,y).log()
```

### Calculate the accuracy

```python
def accuracy(y_hat,y):
    return (y_hat.argmax(axis = 1) == y.astype('float32')).mean().asscalar()
```

```python
accuracy(y_hat,y)
```

```python
def evaluate_accuracy(data_iter,net):
    acc_sum,n = 0.0,0
    for X,y in data_iter:
        y = y.astype('float32')
        acc_sum += (net(X).argmax(axis=1)==y).sum().asscalar()
        n += y.size
    return acc_sum / n
```

```python
evaluate_accuracy(test_iter,net)
```

### Train Model

```python
def train_ch3(net,train_iter,test_iter,loss,num_epochs,batch_size,params=None, lr=None,trainer = None):
    for epoch in range(num_epochs):
        train_l_sum,train_acc_sum,n = 0.0, 0.0 , 0
        for X,y in train_iter:
            with autograd.record():
                y_hat = net(X)
                l = loss(y_hat,y).sum()
            l.backward()
            if trainer is None:
                d2l.sgd(params,lr,batch_size)
            else:
                trainer.step(batch_size)
            y = y.astype('float32')
            train_l_sum += l.asscalar()
            train_acc_sum += (y_hat.argmax(axis = 1) == y).sum().asscalar()
            n += y.size
            test_acc = evaluate_accuracy(test_iter,net)
            print("epoch %d, loss %.4f, train acc %.3f, test acc %.3f" %(epoch +1, train_l_sum/n,train_acc_sum/n,test_acc))
```

```python
train_ch3(net,train_iter,test_iter,cross_entropy,num_epochs,batch_size,[W,b],lr=0.1)
```
